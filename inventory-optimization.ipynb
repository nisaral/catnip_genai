```python
num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')

# Impute numerical columns
imputed_num = pd.DataFrame(
    num_imputer.fit_transform(df[numerical_columns]),
    columns=numerical_columns,
    index=df.index
)

# Impute categorical columns
imputed_cat = pd.DataFrame(
    cat_imputer.fit_transform(df[categorical_columns]),
    columns=categorical_columns,
    index=df.index
)

# Combine imputed numerical and categorical data
df11 = pd.concat([imputed_num, imputed_cat], axis=1)
df11
```

```python
df['year'] = df['Order Date'].dt.year
df['month'] = df['Order Date'].dt.month  
df['day'] = df['Order Date'].dt.day
```

**inventory turnover ratio**

```python
df11["Inventory Turnover Ratio"] = df11["Order Quantity"] / (df11["Current Stock Quantity"] / 2)
```

**safety stock**

```python
df11["Safety Stock"] = 1.65 * (0.2 * df11["Order Quantity"]) * df11["Average Lead Time (days)"].apply(np.sqrt)
```

**Reorder Points (ROP)**

```python
df11["Daily Demand"] = df11["Order Quantity"] / 30
df11["Reorder Point"] = (df11["Daily Demand"] * df11["Average Lead Time (days)"]) + df11["Safety Stock"]
```

**Stock classification (status)**

```python
def classify_stock_status(row):
    if row["Current Stock Quantity"] >= row["Reorder Point"] + row["Safety Stock"]:
        return "Safe"
    elif row["Current Stock Quantity"] < row["Reorder Point"] and row["Current Stock Quantity"] > row["Safety Stock"]:
        return "Reorder"
    else:
        return "Stockout"

df11["Stock Status"] = df11.apply(classify_stock_status, axis=1)
```
